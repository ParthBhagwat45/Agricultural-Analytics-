```python
import numpy as np
import pandas as pd 

df = pd.read_csv("Crop_production.csv")
df.head()
```

```python
df.drop(columns = "Unnamed: 0",inplace = True)
```

```python
df.info()
```

```python
df.duplicated().sum()
```

```python
def soil_type(pH):
    if pH < 7:
        return "Acidic"
    elif pH == 7:
        return "Neutral"
    else :
        return "Basic"


df['Soil_type'] = df['pH'].apply(soil_type)
df
```

```python
df.describe()
```

```python
print((df['Production_in_tons'] == 0).sum())
print((df['Yield_ton_per_hec'] == 0).sum())
print((df['Area_in_hectares'] == 0).sum())
```

```python
def rainfall_level(rainfall):
    if rainfall <= 157:
        return "Low"
    elif rainfall<= 580:
        return "Moderate"
    elif rainfall <= 1111:
        return "High"
    else : 
        return "Very High"


df['Rainfall_level'] = df['rainfall'].apply(rainfall_level)

df['Rainfall_level'].value_counts()
```

```python
def temperature_level(temperature):
   if temperature <= 23.1:
       return "Low"
   elif temperature <= 27.3:
       return "Moderate"
   elif temperature <= 29.3:
       return "High"
   else:
       return "Very High"


df['Temperature_level'] = df['temperature'].apply(temperature_level)

df['Temperature_level'].value_counts()
```

```python
df.to_csv("Crop_production_cleaned.csv",index = False)
```

```python
# -------------------------------
# WEATHER ANALYSIS
# -------------------------------

print("="*60)
print("RAINFALL vs PRODUCTION")
print(df['rainfall'].corr(df['Production_in_tons']))
print()

print("="*60)
print("RAINFALL vs YIELD")
print(df['rainfall'].corr(df['Yield_ton_per_hec']))
print()

print("="*60)
print("TEMPERATURE vs PRODUCTION")
print(df['temperature'].corr(df['Production_in_tons']))
print()

print("="*60)
print("TEMPERATURE vs YIELD")
print(df['temperature'].corr(df['Yield_ton_per_hec']))
print()

print("="*60)
print("PRODUCTION BY RAINFALL LEVEL")
print(df.groupby('Rainfall_level')['Production_in_tons'].mean().sort_values(ascending=False))
print()

print("="*60)
print("YIELD BY RAINFALL LEVEL")
print(df.groupby('Rainfall_level')['Yield_ton_per_hec'].mean().sort_values(ascending=False))
print()

print("="*60)
print("PRODUCTION BY TEMPERATURE LEVEL")
print(df.groupby('Temperature_level')['Production_in_tons'].mean().sort_values(ascending=False))
print()

print("="*60)
print("YIELD BY TEMPERATURE LEVEL")
print(df.groupby('Temperature_level')['Yield_ton_per_hec'].mean().sort_values(ascending=False))
```

```python
# -------------------------------
# SOIL ANALYSIS
# -------------------------------

print("="*60)
print("SOIL TYPE vs PRODUCTION")
print(df.groupby('Soil_type')['Production_in_tons'].mean().sort_values(ascending=False))
print()

print("="*60)
print("SOIL TYPE vs YIELD")
print(df.groupby('Soil_type')['Yield_ton_per_hec'].mean().sort_values(ascending=False))
print()

print("="*60)
print("pH vs PRODUCTION")
print(df['pH'].corr(df['Production_in_tons']))
print()

print("="*60)
print("pH vs YIELD")
print(df['pH'].corr(df['Yield_ton_per_hec']))
print()

print("="*60)
print("N vs PRODUCTION")
print(df['N'].corr(df['Production_in_tons']))
print()

print("="*60)
print("N vs YIELD")
print(df['N'].corr(df['Yield_ton_per_hec']))
print()

print("="*60)
print("P vs PRODUCTION")
print(df['P'].corr(df['Production_in_tons']))
print()

print("="*60)
print("P vs YIELD")
print(df['P'].corr(df['Yield_ton_per_hec']))
print()

print("="*60)
print("K vs PRODUCTION")
print(df['K'].corr(df['Production_in_tons']))
print()

print("="*60)
print("K vs YIELD")
print(df['K'].corr(df['Yield_ton_per_hec']))
```

```python
# -------------------------------
# COMBINED ANALYSIS
# -------------------------------

print("="*60)
print("STATE x CROP -> PRODUCTION")
print(df.groupby(['State_Name','Crop'])['Production_in_tons'].sum().sort_values(ascending=False).head(20))
print()

print("="*60)
print("STATE x CROP -> YIELD")
print(df.groupby(['State_Name','Crop'])['Yield_ton_per_hec'].mean().sort_values(ascending=False).head(20))
print()

print("="*60)
print("STATE x CROP TYPE -> PRODUCTION")
print(df.groupby(['State_Name','Crop_Type'])['Production_in_tons'].sum().sort_values(ascending=False))
print()

print("="*60)
print("STATE x CROP TYPE -> YIELD")
print(df.groupby(['State_Name','Crop_Type'])['Yield_ton_per_hec'].mean().sort_values(ascending=False))
```
